```python
import pandas as pd
import numpy as np
import os
import seaborn as sns
from collections import Counter
from matplotlib import pyplot as plt
```

```python
pwd = os.getcwd()
```

```python
train = pd.read_csv(pwd + '\\train.csv')
test = pd.read_csv(pwd + '\\test.csv')
train.drop(columns=['Survived','PassengerId'],inplace=True)
test.drop(columns='PassengerId',inplace=True)
dataset = train.append(test)
dataset
```

```python
names = []
for i in dataset['Name']:
    names.append(i.split(',')[1].split('.')[0].strip())
Counter(names)
```

```python
dataset['Titles'] = names
Title_Dictionary = {
    "Capt": "Officer",
    "Col": "Officer",
    "Major": "Officer",
    "Jonkheer": "Royalty",
    "Don": "Royalty",
    "Sir" : "Royalty",
    "Dr": "Officer",
    "Rev": "Officer",
    "the Countess":"Royalty",
    "Mme": "Mrs",
    "Mlle": "Miss",
    "Ms": "Mrs",
    "Mr" : "Mr",
    "Mrs" : "Mrs",
    "Miss" : "Miss",
    "Master" : "Master",
    "Lady" : "Royalty"
}
dataset['Titles'] = dataset['Titles'].map(Title_Dictionary)
dataset
```

```python
dataset.groupby([dataset['Sex'],dataset['Pclass'],dataset['Titles']],as_index=False)['Age'].median()
```

```python
pd.pivot_table(dataset, index=['Sex','Pclass','Titles'],values='Age',aggfunc='median',fill_value=0)
```
