```python
import pandas as pd
import numpy as np
import os
import datetime
import plotly.express as px
```

```python
pwd = os.getcwd()
```

# <span style="color:#337ab7">Confirmed</span>

```python
confirmed = pd.read_csv(pwd + '\\time_series_covid_19_confirmed.csv')
confirmed
```

```python
confirmed_melted = confirmed.melt(
    id_vars=list(confirmed.columns)[:4], value_vars=list(confirmed.columns)[4:], var_name='ObservationDate', value_name='Confirmed'
)
confirmed_melted
```

```python
confirmed_melted['ObservationDate'] = pd.to_datetime(confirmed_melted['ObservationDate'])

confirmed_melted_2 = confirmed_melted.groupby(
    [pd.Grouper(key='ObservationDate',freq='M'), 'Province/State', 'Country/Region'], dropna=False
).sum().reset_index()

confirmed_melted_2.drop(columns = ['Lat',"Long"], inplace=True)
confirmed_melted_2
```

```python
output1 = confirmed_melted_2[confirmed_melted_2['ObservationDate'] == '2021-03-31']
output1['ObservationDate'] = output1['ObservationDate'].map(lambda x: x.strftime('%m/%d/%Y'))
output1
```

# <span style="color:#337ab7">Deaths</span>

```python
deaths = pd.read_csv(pwd + '\\time_series_covid_19_deaths.csv')
deaths
```

```python
deaths_melted = deaths.melt(
    id_vars=list(deaths.columns)[:4], value_vars=list(deaths.columns)[4:], var_name='ObservationDate', value_name='Deaths'
)
deaths_melted
```

```python
deaths_melted['ObservationDate'] = pd.to_datetime(deaths_melted['ObservationDate'])

deaths_melted_2 = deaths_melted.groupby(
    [pd.Grouper(key='ObservationDate',freq='M'), 'Province/State', 'Country/Region'], dropna=False
).sum().reset_index()

deaths_melted_2.drop(columns = ['Lat',"Long"], inplace=True)
deaths_melted_2
```

```python
output2 = deaths_melted_2[deaths_melted_2['ObservationDate'] == '2021-03-31']
output2['ObservationDate'] = output2['ObservationDate'].map(lambda x: x.strftime('%m/%d/%Y'))
output2
```

# <span style="color:#337ab7">Recovered</span>

```python
reco = pd.read_csv(pwd + '\\time_series_covid_19_recovered.csv')
reco
```

```python
reco_melted = reco.melt(
    id_vars=list(reco.columns)[:4], value_vars=list(reco.columns)[4:], var_name='ObservationDate', value_name='Recovered'
)
reco_melted
```

```python
reco_melted['ObservationDate'] = pd.to_datetime(reco_melted['ObservationDate'])

reco_melted_2 = reco_melted.groupby(
    [pd.Grouper(key='ObservationDate',freq='M'), 'Province/State', 'Country/Region'], dropna=False
).sum().reset_index()

reco_melted_2.drop(columns = ['Lat',"Long"], inplace=True)
reco_melted_2
```

```python
output3 = reco_melted_2[reco_melted_2['ObservationDate'] == '2021-03-31']
output3['ObservationDate'] = output3['ObservationDate'].map(lambda x: x.strftime('%m/%d/%Y'))
output3
```

# <span style="color:#337ab7">Output Final</span>

```python
on = ['ObservationDate', 'Province/State', 'Country/Region']
output_final = pd.merge(left=output1, right=output2, how='inner', left_on=on, right_on=on)
output_final = pd.merge(left=output_final, right=output3, how='outer', left_on=on, right_on=on)
output_final['Province/State'].fillna('-', inplace=True)
output_final
```

```python
output_final.groupby('Country/Region').sum()
```
